Below is code with a link to a happy or sad dataset which contains 80 images, 40 happy and 40 sad. 
Create a convolutional neural network that trains to 100% accuracy on these images,  which cancels training upon hitting training accuracy of >.999

Hint -- it will work best with 3 convolutional layers.

```
%tensorflow_version 2.x
import tensorflow as tf
print(tf.__version__)
import os
import zipfile


DESIRED_ACCURACY = 0.999

!wget --no-check-certificate \
    "https://storage.googleapis.com/laurencemoroney-blog.appspot.com/happy-or-sad.zip" \
    -O "/tmp/happy-or-sad.zip"

zip_ref = zipfile.ZipFile("/tmp/happy-or-sad.zip", 'r')
zip_ref.extractall("/tmp/h-or-s")
zip_ref.close()

class myCallback(tf.keras.callbacks.Callback):
  # Your Code
  def on_epoch_end(self, epoch, logs={}):
    if logs['accuracy']>DESIRED_ACCURACY:
      print('\n Reached %0.1f%% training accuracy. Training converged!' %(DESIRED_ACCURACY*100))
      self.model.stop_training=True

callback = myCallback()
```

```
# This Code Block should Define and Compile the Model
model = tf.keras.models.Sequential([
    # Your Code Here
    tf.keras.layers.Conv2D(12, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Conv2D(12, (3,3), activation='relu'),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Conv2D(12, (3,3), activation='relu'),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

from tensorflow.keras.optimizers import RMSprop

model.compile(
    # Your Code Here
    optimizer=RMSprop(lr=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
    )
```

```
# This code block should create an instance of an ImageDataGenerator called train_datagen 
# And a train_generator by calling train_datagen.flow_from_directory

from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1/255) # Your Code Here

train_generator = train_datagen.flow_from_directory(
        # Your Code Here
        '/tmp/h-or-s/',
        target_size=(150, 150),
        batch_size=10,
        class_mode='binary'
        )

# Expected output: 'Found 80 images belonging to 2 classes'
```

```
# This code block should call model.fit_generator and train for
# a number of epochs. 
history = model.fit_generator(
      # Your Code Here
      train_generator,
      steps_per_epoch=train_generator.n//train_generator.batch_size,
      epochs=15,
      callbacks=[callback],
      verbose=1 #,
      # validation_data=validation_generator
      # validation_steps=validation_generator.n//validation_generator.batch_size
      )
    
# Expected output: "Reached 99.9% accuracy so cancelling training!""
```
